```python
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("WEATHER_API_KEY")
print("Loaded API key:", bool(API_KEY))
```

```python
import requests
import json

def get_weather_data(city):
    """Fetch weather data for specified city"""
    url = f"https://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}"
    response = requests.get(url)
    return response.json()

def save_results(data, filename="weather_data.json"):
    """Save processed results to a file"""
    with open(filename, "w") as f:
        json.dump(data, f, indent=2)

def weather_agent(city):
    data = get_weather_data(city)
    # Process data here
    save_results(data)
    return f"Weather in {city}: {data['current']['temp_c']}°C"
```

```python
# Example usage:
city_name = "London"  # <-- edit this to any city you want
result = weather_agent(city_name)
print(result)
```
